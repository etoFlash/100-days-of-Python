For another text editors: plt.show()

```python
import numpy as np
```

```python
a = np.linspace(0, 5, 10)
b = a ** 3
```

```python
a
```

```python
b
```

### Functional methos

```python
plt.plot(a, b)
plt.xlabel('a array')
plt.ylabel('b array')
plt.title('Plot for a and b arrays')
```

```python
plt.subplot(1, 2, 1)
plt.plot(a, b, 'r')
plt.subplot(1, 2, 2)
plt.plot(b, a, 'b')
```

### Object oriented method

```python
figure = plt.figure()
axes = figure.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(a, b)
axes.set_xlabel('a array values')
axes.set_ylabel('b array values')
axes.set_title('Some title')
```

```python
figure1 = plt.figure()
axes1 = figure1.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = figure1.add_axes([0.3, 0.6, 0.5, 0.4])
```

```python
figure1 = plt.figure()
axes1 = figure1.add_axes([0.0, 0.0, 1, 1])
axes2 = figure1.add_axes([0.3, 0.6, 0.5, 0.4])
```

```python
figure1 = plt.figure()
axes1 = figure1.add_axes([0.0, 0.0, 1, 1])
axes2 = figure1.add_axes([0.5, 0.0, 0.5, 0.4])
```

```python
figure1 = plt.figure()
axes1 = figure1.add_axes([0.0, 0.0, 1, 1])
axes2 = figure1.add_axes([0.1, 0.4, 0.5, 0.4])

axes1.plot(a, b)
axes2.plot(b, a)

axes1.set_xlabel('a array values')
axes1.set_ylabel('b array values')
axes1.set_title('Axes1 title')

axes2.set_xlabel('b array values')
axes2.set_ylabel('a array values')
axes2.set_title('Axes2 title')
```

```python
figure1 = plt.figure()
axes1 = figure1.add_axes([0.0, 0.0, 1, 1])
axes1.plot(a, b)
```

# Part 2

```python
figure, axes = plt.subplots(nrows=2, ncols=4)
plt.tight_layout()
# axes.plot(a, b)
```

```python
figure, axes = plt.subplots(nrows=1, ncols=2)
for ax in axes:
    ax.plot(a, b)
```

```python
axes
```

```python
figure, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(a, b)
axes[0].set_xlabel('a array values')
axes[0].set_ylabel('b array values')
axes[0].set_title('Axes[0] title')

axes[1].plot(b, a)
axes[1].set_xlabel('b array values')
axes[1].set_ylabel('a array values')
axes[1].set_title('Axes[1] title')
```

## Figure size, aspect ratio and dpi

```python
figure = plt.figure(figsize=(7, 2))
axes = figure.add_axes([0, 0, 1, 1])
axes.plot(a, b)
```

```python
figure, axes = plt.subplots(nrows=2, ncols=1, figsize=(7, 2))
axes[0].plot(a, b)
axes[1].plot(b, a)
plt.tight_layout()
```

```python
figure = plt.figure()
axes = figure.add_axes([0, 0, 1, 1])
axes.plot(a, b, label='a, b plot')
axes.plot(b, a, label='b, a plot')
# https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.legend.html
axes.legend(loc=(0.1, 0.1))
```
